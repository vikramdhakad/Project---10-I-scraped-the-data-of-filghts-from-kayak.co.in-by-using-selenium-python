## Import required python libraries

```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.action_chains import ActionChains
from time import sleep
```

```python
chrome_options = Options()
chrome_options.add_experimental_option("detach",True)

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),options=chrome_options)
driver.get("https://www.kayak.co.in/flights/DEL-GOI/2024-05-30?sort=bestflight_a#default")
sleep(3)
driver.maximize_window()
```

```python
for n in range(1,9):
    nt = driver.find_element(By.XPATH,"//div[@class='ULvh-button show-more-button']")
    nt.click()
    sleep(5)

driver.find_element(By.TAG_NAME,'body').send_keys(Keys.CONTROL + Keys.HOME)
```

```python
name = []
price = []
time = []
airports = []
tt = []
ft = []

pages = driver.find_elements(By.XPATH,"//div[@class='nrc6-content-section']")
for page in pages:
    page.click()
    sleep(3)
    try:
        flight_name = driver.find_element(By.XPATH,"//div[@class='NxR6-plane-details']/div[1]")
        name.append(flight_name.text)
    except:
        name.append(" ")

    try:
        flight_price = driver.find_element(By.XPATH,"//div[@class='jnTP-display-price']")
        price.append(flight_price.text)
    except:
        price.append(" ")

    try:
        flight_timing = driver.find_element(By.XPATH,"//div[@class='NxR6-time']")
        time.append(flight_timing.text)
    except:
        time.append(" ")

    try:
        airport = driver.find_element(By.XPATH,"//div[@class='c2x94-title']/span[1]")
        airports.append(airport.text)
    except:
        airports.append(" ")

    try:
        time_take = driver.find_element(By.XPATH,"//span[@class='NxR6-duration']")
        tt.append(time_take.text)
    except:
        tt.append(" ")

    try:
        flight_type = driver.find_element(By.XPATH,"//div[@class='c2x94-sub-title']")
        ft.append(flight_type.text)
    except:
        ft.append(" ")
    driver.back()
```

```python
data = pd.DataFrame()
data['Flight Name'] = name
data['Ticket Price'] = price
data['Flight timming'] = time
data['Airport to Airport'] = airports
data['Time Take'] = tt
data['Flight stops'] = ft

data 
```

```python
data.to_csv("Flight_Kayak.csv")
```
